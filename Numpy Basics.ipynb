```python
import cv2

facedetector = cv2.CascadeClassifier('face_template.xml')
```

```python
faces = facedetector.detectMultiScale(img,1.2,5)
```

```python
for f in faces:
    x,y,w,h = f
    green = (0,255,0)
    cv2.rectangle(img,(x,y),(x+w,y+h),green,5)

plt.imshow(img)
plt.imsave('test.jpg',img)
plt.show()
```

```python
cropped_faces = []

for f in faces:
    x,y,w,h = f
    green = (0,255,0)
    cv2.rectangle(img,(x,y),(x+w,y+h),green,5)
    cropped_faces.append(img[y:y+h,x:x+w])
    

plt.imshow(img)
plt.imsave('test.jpg',img)
plt.show()
```

```python
new_list = []
for cf in cropped_faces:
    cf_resize = cv2.resize(cf,(100,100))
    new_list.append(cf_resize)
    #plt.imshow(cf_resize)
    #plt.show()
```

```python
all_f = np.hstack(new_list)
plt.imshow(all_f)
plt.show()
```
