```python
import cv2
import numpy as np

camera = cv2.VideoCapture(0)
# Create a Face Detector Object
facedetector = cv2.CascadeClassifier('../Day-1/face_template.xml')

name = input("Enter the name of person getting clicked ")
pics = []
cnt = 0


while True:
    b,img = camera.read()
    
    if b==False:
        continue
    # Detect Faces
    faces = facedetector.detectMultiScale(img,1.2,5)
    
    # No face is detected
    if(len(faces)==0):
        continue

    
    # Draw bounding box around face
    face = faces[0]
    x,y,w,h = faces[0]
    green = (0,255,0)
    cv2.rectangle(img,(x,y),(x+w,y+h),green,5)
    
    # Crop and Save that face pic in pictures
    
    cropped_face = img[y:y+h,x:x+w]
    cv2.imshow("cropped face",cropped_face)
    cnt += 1
    
    # Save every 10th frame
    if cnt%10==0:
        print("Saving picture ",cnt/10)
        cropped_face = cv2.resize(cropped_face,(100,100))
        pics.append(cropped_face)
    
    if cnt/10==20:
        break
    
    
    # Show the New Image
    cv2.imshow("Title",img)
    #Add some delay 1 ms between 2 frames
    cv2.waitKey(1)
    
camera.release()
cv2.destroyAllWindows()

pics = np.array(pics)
print(pics.shape)

np.save(name+".npy",pics)
print("Images saved to disk")
```
